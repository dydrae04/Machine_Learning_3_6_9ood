```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
```

```python
import pandas as pd

file_path = "data/universal_top_spotify_songs.csv"

df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')

print(df.shape)
df.head()
```

```python
features = [
    'danceability','energy','loudness','speechiness',
    'acousticness','instrumentalness','liveness','valence','tempo'
]
df = df.dropna(subset=features)

top5 = df['country'].value_counts().head(5).index.tolist()
print("상위 5개국:", top5)

K = 5   # Top5 기준
results = []

for country in top5:
    sub = df[df['country'] == country].copy()
    grouped = sub.groupby('spotify_id')[features + ['daily_rank']].mean().reset_index()

    topK_ids = sub.loc[sub['daily_rank'] <= K, 'spotify_id'].unique()
    grouped['TopK'] = grouped['spotify_id'].isin(topK_ids).astype(int)

    if grouped['TopK'].nunique() < 2:
        continue

    # train/test 분리
    spotify_ids = grouped['spotify_id'].unique()
    train_ids, test_ids = train_test_split(
        spotify_ids, test_size=0.3, random_state=0
    )

    train_data = grouped[grouped['spotify_id'].isin(train_ids)]
    test_data  = grouped[grouped['spotify_id'].isin(test_ids)]

    X_train, y_train = train_data[features], train_data['TopK']
    X_test, y_test   = test_data[features],  test_data['TopK']

    # n_estimators 
    scores = []
    for n_trees in range(10, 500, 5):
        forest_model = RandomForestClassifier(n_estimators=n_trees, random_state=0)
        forest_model.fit(X_train, y_train)
        preds = forest_model.predict(X_test)
        acc = accuracy_score(y_test, preds)
        scores.append(acc)
        print('> %d, acc: %.3f' % (n_trees, acc))

    best_n = list(range(10, 500, 5))[scores.index(max(scores))]
    best_acc = max(scores)
    print(f" [{country}] 최적 n_estimators = {best_n}, Accuracy = {best_acc:.3f}")

    #학습
    final_forest_model = RandomForestClassifier(n_estimators=best_n, random_state=0)
    final_forest_model.fit(X_train, y_train)
    y_pred = final_forest_model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    print(f" [{country}] 최종 모델 Accuracy: {acc:.3f}")

    # 결과 
    results.append({'Country': country, 'Best_n_estimators': best_n, 'Accuracy': round(acc, 3)})

    # 시각화
    plt.figure(figsize=(6, 4))
    plt.plot(range(10, 500, 5), scores, color='teal')
    plt.title(f"{country} — n_estimators별 Accuracy 변화")
    plt.xlabel("n_estimators")
    plt.ylabel("Accuracy")
    plt.grid(True)
    plt.show()

result_df = pd.DataFrame(results)
print("\n 국가별 최적 n_estimators, Accuracy 결과")
print(result_df)
```
