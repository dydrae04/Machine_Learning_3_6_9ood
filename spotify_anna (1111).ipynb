```python
import pandas as pd

file_path = "data/universal_top_spotify_songs.csv"

df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')

print(df.shape)
df.head()
```

```python
# 음악 특성과 관련 없는 변수 제거
drop_cols = ['artists','album_name','album_release_date','spotify_id']

df = df.drop(columns=drop_cols)
df
```

```python
# 날짜 필터링: 2024-01-01 ~ 2024-12-31
df['snapshot_date'] = pd.to_datetime(df['snapshot_date'], errors='coerce')
df = df[(df['snapshot_date'] >= "2024-01-01") & (df['snapshot_date'] <= "2024-12-31")]
df
```

```python
# 국가 null확인
df[df['country'].isnull()]
```

```python
# 국가 null 제거
df=df.dropna(subset=['country'])
df
```

```python
# 국가종류 확인
all_countries = df['country'].unique()

print("데이터셋에 포함된 모든 국가 코드:")
print(all_countries)
```

```python
continent_map = {
    'Asia': [
        'AE', 'IL', 'IN', 'ID', 'JP', 'KR', 'KZ', 'MY', 'PH', 'SG', 'TH', 'TW', 'VN', 'HK', 'SA', 'TR'
    ],
    'Europe': [
        'AT', 'BE', 'BG', 'BY', 'CH', 'CZ', 'DE', 'DK', 'EE', 'ES', 'FI', 'FR', 'GR', 'HU', 'IE', 'IS',
        'IT', 'LT', 'LU', 'LV', 'NL', 'NO', 'PL', 'PT', 'RO', 'SE', 'SK', 'UA', 'GB'
    ],
    'North America': [
        'CA', 'CR', 'DO', 'GT', 'HN', 'MX', 'NI', 'PA', 'SV', 'US'
    ],
    'South America': [
        'AR', 'BO', 'BR', 'CL', 'CO', 'EC', 'PE', 'PY', 'UY', 'VE'
    ],
    'Africa': [
        'EG', 'MA', 'NG', 'ZA'
    ],
    'Oceania': [
        'AU', 'NZ'
    ]
}
```

```python
def get_continent(code):
    for continent, countries in continent_map.items():
        if code in countries:
            return continent
    return 'Unknown'
```

```python
df['continent'] = df['country'].apply(get_continent)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

# X: 음악 특성 / y: 대륙
features = [
    'danceability','energy','loudness','speechiness',
    'acousticness','instrumentalness','liveness','valence','tempo'
]
X = df[features]
y = df['continent']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,          # 20%를 테스트 데이터로
    random_state=42,        # 재현 가능한 분할
    stratify=y              # 대륙별 비율 동일하게
)

print("학습용 데이터:", X_train.shape)
print("테스트용 데이터:", X_test.shape)
print("\n대륙 비율 확인:")
print("Train:\n", y_train.value_counts(normalize=True).round(3))
print("\nTest:\n", y_test.value_counts(normalize=True).round(3))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
train_scores, test_scores = list(),list()
```

```python
for i in range(1, 51):
    DT = DecisionTreeClassifier(max_depth=i, random_state=0)
    DT.fit(X_train, y_train)

    y_train_pred = DT.predict(X_train)
    y_test_pred = DT.predict(X_test)

    train_acc = accuracy_score(y_train, y_train_pred)
    test_acc = accuracy_score(y_test, y_test_pred)

    train_scores.append(train_acc)
    test_scores.append(test_acc)

    print(f'>{i}, train: {train_acc:.3f}, test: {test_acc:.3f}')
```

```python
Tree = DecisionTreeClassifier(max_depth=27, random_state=0)
Tree.fit(X_train, y_train)
```

```python
y_pred_dt = Tree.predict(X_test)
temp_acc = accuracy_score(y_test, y_pred_dt)
print(f'의사결정나무 정확도: {temp_acc:.4f}')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF=RandomForestClassifier(n_estimators=50,random_state=0)
RF.fit(X_train, y_train)
y_pred_rf=RF.predict(X_test)

rf_acc=accuracy_score(y_test,y_pred_rf)
print(f'랜덤포레스트 정확도: {rf_acc:4f}')
```
